```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('penguins')
```

```python
df.head()
```

```python
df["species"].value_counts
```

```python
df["island"].value_counts()
```

```python
sns.set_style("darkgrid")
sns.scatterplot(data=df,x="flipper_length_mm",y="body_mass_g",hue="island",style="sex")
```

```python
sns.stripplot(data=df,x="species",y="body_mass_g",hue="island",dodge=True)
```

```python

sns.swarmplot(data=df,x="species",y="body_mass_g",hue="island")
```

```python
sns.set_style("ticks")
sns.histplot(data=df,x="body_mass_g",hue="sex")
```

```python
sns.regplot(data=df,x="body_mass_g",y="flipper_length_mm")
```

```python
sns.set_style("whitegrid")
sns.lineplot(data=df,x="body_mass_g",y="flipper_length_mm",hue="sex")
```

```python
sns.jointplot(data=df,x="body_mass_g",y="flipper_length_mm",hue = "sex",kind="kde")
```

```python
sns.set_style("white")
sns.set_context("notebook")
sns.barplot(data=df,y="body_mass_g",x="species",palette=["red","black"])
```

```python
sns.countplot(data=df,x="species")
```

```python
sns.set_style("whitegrid")
sns.boxplot(data=df,x="species",y="body_mass_g",hue="sex")
```

```python

sns.kdeplot(data=df,y="flipper_length_mm",x="body_mass_g",fill=True)
```
